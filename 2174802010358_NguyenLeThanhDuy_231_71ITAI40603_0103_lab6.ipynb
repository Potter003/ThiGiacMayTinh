```python
import cv2
import matplotlib.pyplot as nltd

img = cv2.imread("D:\ThiGiacMayTinh\LAB6/stop.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

stop_data = cv2.CascadeClassifier('D:\ThiGiacMayTinh\LAB6/stop_data.xml')

found = stop_data.detectMultiScale(img_gray, minSize=(20, 20))
amount_found = len(found)

if amount_found != 0:
    for (x, y, width, height) in found:
        cv2.rectangle(img_rgb, (x, y), (x + width, y + height), (0, 255, 0), 5)

nltd.imshow(img_rgb)
nltd.show()


import cv2
from matplotlib import pyplot as nltd
   
img = cv2.imread("D:\ThiGiacMayTinh/LAB6/highschool.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
   
face_data = cv2.CascadeClassifier('D:\ThiGiacMayTinh/LAB6/haarcascade_frontalface_default.xml')

found = face_data.detectMultiScale(img_gray, minSize =(20, 20))

amount_found = len(found)
   
if amount_found != 0:
       
    for (x, y, width, height) in found:
           
        cv2.rectangle(img_rgb, (x, y), 
                      (x + width, y + height), 
                      (0, 255, 0), 2)
        
nltd.imshow(img_rgb)
nltd.show()

import cv2
import matplotlib.pyplot as nltd

img = cv2.imread("D:\ThiGiacMayTinh\LAB6/highschool.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

smile_data = cv2.CascadeClassifier('D:\ThiGiacMayTinh\LAB6/haarcascade_mcs_fullface.xml')

found = smile_data.detectMultiScale(img_gray, minSize=(20, 20))
amount_found = len(found)

if amount_found != 0:
    for (x, y, width, height) in found:
        cv2.rectangle(img_rgb, (x, y), (x + width, y + height), (0, 255, 0), 5)

nltd.imshow(img_rgb)
nltd.show()
```
